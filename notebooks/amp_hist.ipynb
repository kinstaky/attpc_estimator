# Amplitude of found peaks

```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

%matplotlib widget

workspace = Path("/data/rcnp2025/attpc_estimator")
hist_file = workspace / "run_1056_amp.npy"
```

```python
payload = np.load(hist_file, allow_pickle=True)
fig, ax = plt.subplots(figsize=(10, 6), constrained_layout=True)
ax.stairs(payload)
ax.set_title("Amplitude of found peaks")
ax.set_xlabel("Amplitude")
ax.set_ylabel("Counts")
ax.set_yscale('log')
```
